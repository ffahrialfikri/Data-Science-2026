```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, matplotlib.gridspec as gridspec
import seaborn as sns


# ── 1. LOAD & INSPECT DATASET ──────────────────────────────────────────
df = sns.load_dataset('diamonds')
print(df.shape, df.dtypes)
print(df.describe().round(2))

# ── 2. SETUP FIGURE LAYOUT ─────────────────────────────────────────────
fig = plt.figure(figsize=(16, 12))
fig.suptitle('Dashboard Analisis Dataset Diamonds',
fontsize=18, fontweight='bold', y=0.98)

# Grid 2 x 2 subplot
gs = gridspec.GridSpec(2, 2, figure=fig,
hspace=0.4, wspace=0.35)
ax1 = fig.add_subplot(gs[0, 0]) # kiri atas
ax2 = fig.add_subplot(gs[0, 1]) # kanan atas
ax3 = fig.add_subplot(gs[1, 0]) # kiri bawah
ax4 = fig.add_subplot(gs[1, 1]) # kanan bawah

# ── 3. GRAFIK 1: Histogram of Carat ────────────────────────────────────
sns.histplot(data=df, x='carat', kde=True, ax=ax1, color='skyblue')
ax1.set_title('Distribusi Berat Karat')
ax1.set_xlabel('Carat'); ax1.set_ylabel('Count')
ax1.spines[['top','right']].set_visible(False)

# ── 4. GRAFIK 2: Countplot of Cut Quality ──────────────────────────────
sns.countplot(data=df, x='cut', ax=ax2, palette='viridis', order=df['cut'].value_counts().index)
ax2.set_title('Distribusi Kualitas Potongan (Cut)')
ax2.set_xlabel('Cut'); ax2.set_ylabel('Count')
ax2.spines[['top','right']].set_visible(False)

# ── 5. GRAFIK 3: Boxplot of Price by Color ─────────────────────────────
sns.boxplot(data=df, x='color', y='price', ax=ax3, palette='magma', order=df['color'].value_counts().index)
ax3.set_title('Distribusi Harga Berdasarkan Warna')
ax3.set_xlabel('Color'); ax3.set_ylabel('Price')
ax3.spines[['top','right']].set_visible(False)

# ── 6. GRAFIK 4: Scatter Plot of Carat vs. Price ───────────────────────
sns.scatterplot(data=df, x='carat', y='price', hue='cut', size='depth', sizes=(20, 200), alpha=0.6, ax=ax4)
sns.regplot(data=df, x='carat', y='price', scatter=False, color='gray', ax=ax4)
ax4.set_title('Hubungan Berat Karat dan Harga (Ukuran = Kedalaman)')
ax4.set_xlabel('Carat'); ax4.set_ylabel('Price')
ax4.spines[['top','right']].set_visible(False)

# ── 7. EKSPOR ──────────────────────────────────────────────────────────
plt.savefig('dashboard_diamonds.png', dpi=150, bbox_inches='tight')
plt.show()
print('Dashboard tersimpan sebagai dashboard_diamonds.png')
```
