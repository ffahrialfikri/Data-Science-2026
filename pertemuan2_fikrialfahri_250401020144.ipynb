```python
#Nama: Fikri Alfahri NIM:250401020144 Kelas: IF405

##Python — List Operations
# Membuat list
data = [10, 20, 5, 15, 30] # Modified list to contain only numbers for sorting
# Akses & slicing
print(data[0]) # → 10
print(data[-1]) # → 30 (dari belakang)
print(data[1:3]) # Reverted slice to a more common usage to avoid empty list output
# Manipulasi
data.append(99) # tambah di akhir
data.insert(1, 55) # sisipkan di posisi 1
data.remove(20) # hapus nilai pertama = 20
data.sort() # urutkan (jika tipe sama)

print(data)

print("=" * 40)



##Python — Dictionary Operations
mahasiswa = {
'nama': 'Budi Santoso',
'usia': 20,
'kota': 'Jakarta',
'nim': '2024001',
'prodi': 'Informatika',
'ipk': 3.85
}
# Akses nilai
print(mahasiswa['nama']) # → 'Budi Santoso'
print(mahasiswa.get('usia', 0)) # default jika key tak ada
# Update dan hapus
mahasiswa['angkatan'] = 2024 # tambah key baru
del mahasiswa['nim'] # hapus key

print("=" * 40)




##Python — Membuat NumPy Array
import numpy as np
# Dari list Python
a1D = np.array([1, 2, 3, 4, 5]) # 1D
a2D = np.array([[1,2,3],[4,5,6]]) # 2D
# Array dengan nilai inisial
zeros = np.zeros((3, 4)) # nol semua
ones = np.ones((2, 3)) # satu semua
eye = np.eye(4) # matriks identitas
full = np.full((2, 2), 7) # nilai konstan
# Sequence
arange = np.arange(0, 20, 2) # 0,2,4,...18
linsp = np.linspace(0, 1, 11) # 11 titik merata
# Array acak (reproducible)
rng = np.random.default_rng(seed=42)
rand_int = rng.integers(0, 100, (3, 4))
rand_flt = rng.normal(loc=0, scale=1, size=(100,))

# Menampilkan beberapa array NumPy yang telah dibuat
print('a1D Array:\n', a1D)
print('\na2D Array:\n', a2D)
print('\nZeros Array (3x4):\n', zeros)
print('\nOnes Array (2x3):\n', ones)
print('\nIdentity Matrix (4x4):\n', eye)
print('\nFull Array (2x2, value 7):\n', full)
print('\nArange (0-20, step 2):\n', arange)
print('\nLinspace (0-1, 11 points):\n', linsp)
print('\nRandom Integers (3x4):\n', rand_int)
print('\nRandom Floats (normal distribution, first 5):\n', rand_flt[:5])

print("=" * 40)



##Python — Pandas Series
import pandas as pd
# Series dari list — index default 0,1,2...
s1 = pd.Series([85, 92, 78, 95, 88])
# Series dengan index kustom
nilai = pd.Series(
[85, 92, 78, 95],
index=['Alice','Bob','Caca','Dina']
)
print(nilai['Bob']) # → 92
print(nilai[nilai>80]) # filter
print(nilai.mean()) # → 87.5
print(nilai.describe()) # statistik lengkap

print("=" * 40)




##Python — Pandas DataFrame
# Membuat DataFrame dari dictionary
data_df = {
    'Nama': ['Andi', 'Budi', 'Cici', 'Dedi'],
    'Usia': [23, 22, 24, 23],
    'Kota': ['Bandung', 'Jakarta', 'Surabaya', 'Medan']
}
df_example = pd.DataFrame(data_df)
print('DataFrame yang baru dibuat:\n', df_example)

# Mengakses kolom
print('\nKolom "Nama":\n', df_example['Nama'])

# Filter baris
print('\nFilter usia > 22:\n', df_example[df_example['Usia'] > 22])

print("=" * 40)



##Python — Membuat Pandas DataFrame (Lanjutan)
# Membuat DataFrame dari dictionary
data = {
'nama': ['Ali','Budi','Caca','Dina'],
'usia': [22, 25, 21, 23],
'kota': ['Jakarta','Bandung','Surabaya','Medan'],
'ipk': [3.8, 3.5, 3.9, 3.7]
}
df = pd.DataFrame(data)
# Informasi dasar
print('Shape DataFrame:', df.shape) # → (4, 4)
print('\nDtypes DataFrame:\n', df.dtypes) # tipe data per kolom
print('\nColumns DataFrame:', df.columns) # nama-nama kolom

# Mengakses baris berdasarkan indeks (loc untuk label, iloc untuk posisi)
print('\nBaris dengan index 1 (iloc):\n', df.iloc[1])

# Menambah kolom baru (misal: 'status_ipk' berdasarkan 'ipk')
df['status_ipk'] = np.where(df['ipk'] >= 3.7, 'Sangat Baik', 'Baik')
print('\nDataFrame setelah menambah kolom "status_ipk":\n', df)

# Statistik deskriptif untuk kolom numerik
print('\nStatistik deskriptif untuk kolom numerik:\n', df.describe())

print("=" * 40)




##Python — pd.read_csv()
# Memuat dari file lokal (Ini akan error jika file tidak ada)
# df = pd.read_csv('data/penjualan.csv')

# Memuat dari URL (dataset publik)
# Contoh URL dataset publik dari UCI Machine Learning Repository
url_example = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# Memuat dengan parameter penting, sesuaikan dengan struktur CSV Anda
df_url = pd.read_csv(url_example, header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

print('\nDataFrame dari URL:\n')
print(df_url.head()) # 5 baris pertama
print('\nInfo DataFrame dari URL:\n')
df_url.info() # tipe data & missing
print('\nStatistik deskriptif DataFrame dari URL:\n')
print(df_url.describe()) # statistik deskriptif

# Parameter penting (Contoh dengan placeholder, ini tidak akan berjalan tanpa file 'data.csv')
# df = pd.read_csv('data.csv',
# sep=';', # pemisah kolom (default ',')
# encoding='utf-8', # encoding file
# index_col='ID', # kolom sebagai index
# parse_dates=['tgl'],# parse kolom tanggal
# na_values=['N/A','none'] # nilai kosong
# )
# Eksplorasi standar setelah load
# df.head() # 5 baris pertama
# df.info() # tipe data & missing
# df.describe() # statistik deskriptif


print("=" * 40)




##Python — Boolean Indexing
# Contoh: Dataset Titanic
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(titanic_url)

print('Head of Titanic DataFrame:\n', df.head())

# Filter satu kondisi
penumpang_selamat = df[df['Survived'] == 1]
print('\nPenumpang yang Selamat (pertama 5):\n', penumpang_selamat.head())

penumpang_wanita = df[df['Sex'] == 'female']
print('\nPenumpang Wanita (pertama 5):\n', penumpang_wanita.head())

# Filter kondisi ganda (& = AND, | = OR)
wanita_kelas1 = df[(df['Sex']=='female') &
(df['Pclass']==1)]
print('\nWanita di Kelas 1 (pertama 5):\n', wanita_kelas1.head())

# Seleksi kolom tertentu
print('\nKolom "Name", "Age", "Sex", "Survived":\n', df[['Name','Age','Sex','Survived']].head())

# Menambahkan contoh filter rentang usia
penumpang_dewasa_muda = df[(df['Age'] >= 18) & (df['Age'] <= 35)]
print('\nPenumpang Dewasa Muda (usia 18-35, pertama 5):\n', penumpang_dewasa_muda.head())


print("=" * 40)



##Python — Groupby & Agregasi
# Pola dasar: groupby → aggregation
df.groupby('Pclass')['Survived'].mean()
# Agregasi multiple kolom & fungsi
hasil = df.groupby('Pclass').agg(
total = ('PassengerId', 'count'),
selamat = ('Survived', 'sum'),
pct_selamat= ('Survived', 'mean'),
rata_usia = ('Age', 'mean'),
max_fare = ('Fare', 'max')
)
hasil = hasil.round(3)
print(hasil)
# GroupBy multi-kolom
print('\nGroupby multi-kolom (Pclass dan Sex):\n')
print(df.groupby(['Pclass','Sex'])['Survived']\
.mean().unstack().round(3))

# Visualisasi persentase selamat berdasarkan kelas penumpang
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(x=hasil.index, y=hasil['pct_selamat'], palette='viridis')
plt.title('Persentase Penumpang Selamat Berdasarkan Kelas')
plt.xlabel('Kelas Penumpang (Pclass)')
plt.ylabel('Persentase Selamat')
plt.ylim(0, 1) # Persentase antara 0 dan 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


##SELESAI

##RANGKUMAN
# Dalam notebook ini, kita telah melakukan serangkaian latihan untuk memahami dasar-dasar dan operasi lanjutan dalam analisis data menggunakan Python. Berikut adalah poin-poin utamanya:

# Apa yang Dipelajari:
# Dasar-dasar Python: Kita memulai dengan operasi dasar pada struktur data List dan Dictionary, mencakup pembuatan, akses, slicing, penambahan, penghapusan, dan pengurutan elemen.
# NumPy: Memahami cara membuat dan memanipulasi array NumPy satu dimensi dan multi-dimensi, serta fungsi-fungsi untuk membuat array dengan nilai inisial (zeros, ones, eye, full) dan sequence (arange, linspace), serta array acak.
# Pandas Series: Belajar membuat objek Series dari list, menggunakan indeks kustom, melakukan filtering, dan mendapatkan statistik deskriptif dasar.
# Pandas DataFrame: Mempelajari cara membuat DataFrame dari dictionary, mengakses kolom, dan memfilter baris berdasarkan kondisi. Kemudian dilanjutkan dengan eksplorasi lebih dalam seperti memeriksa shape, dtypes, columns, mengakses baris dengan iloc, menambah kolom baru berdasarkan logika, dan mendapatkan statistik deskriptif seluruh DataFrame.
# Membaca Data CSV: Menggunakan pd.read_csv() untuk memuat data dari URL publik, serta memahami parameter penting untuk membaca data dari file CSV (separator, encoding, index kolom, parse tanggal, nilai kosong).
# Boolean Indexing: Teknik memfilter DataFrame menggunakan kondisi Boolean, baik tunggal maupun ganda, untuk menyeleksi baris data yang spesifik (misalnya, penumpang selamat atau kelompok usia tertentu).
# Groupby & Agregasi: Menerapkan operasi groupby untuk mengelompokkan data berdasarkan satu atau beberapa kolom, kemudian melakukan agregasi dengan berbagai fungsi (count, sum, mean, max) untuk mendapatkan insight seperti persentase survival, rata-rata usia, dan lain-lain. Kita juga belajar cara melakukan groupby multi-kolom dan menampilkan hasilnya dengan unstack().
# Temuan Utama (dari Dataset Titanic):
# Terdapat perbedaan signifikan dalam tingkat kelangsungan hidup antar kelas penumpang (Pclass). Kelas 1 memiliki tingkat kelangsungan hidup tertinggi, diikuti oleh Kelas 2, dan terendah di Kelas 3.
# Jenis kelamin juga menjadi faktor yang sangat dominan, di mana wanita memiliki tingkat kelangsungan hidup yang jauh lebih tinggi dibandingkan pria di semua kelas penumpang.
# Visualisasi menunjukkan bahwa bahkan di kelas terendah (Kelas 3), persentase kelangsungan hidup wanita masih lebih tinggi daripada pria di kelas tertinggi (Kelas 1).
# Tools/Library yang Digunakan:
# numpy: Untuk komputasi numerik dan operasi array.
# pandas: Untuk manipulasi dan analisis data, terutama dengan struktur data Series dan DataFrame.
# matplotlib.pyplot: Untuk membuat visualisasi statis, seperti bar plot.
# seaborn: Untuk membuat visualisasi statistik yang lebih menarik dan informatif, sering digunakan bersama matplotlib.
```
